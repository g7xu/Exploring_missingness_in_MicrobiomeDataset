```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_json('data/hmdb_microbe_metabolites.json')
data
```

```python
hmdb = data[['_id', 'name', 'chemical_formula', 'description', 'status', 'average_mw', 'monoisotopic_mw', 'state', 'sources']]
hmdb = hmdb.assign(inchikey = data['xrefs'].apply(lambda x : x['inchikey']))
hmdb
```

```python
pd.isna(np.NaN)
```

```python
def atomiclize_col(col):
    '''
    expand any columns with complex dictionaries
    ---
    col: takes in a series where each entry is a dictionary
    ---
    return a DataFrame with expanded 
    '''
    key_collection = set()
    for i in col.values:
        if not pd.isna(i):
            key_collection = key_collection.union(set(i.keys()))
    outputDF = pd.DataFrame()
    for coln in key_collection:
        outputDF[col.name+ '_' + coln] = col.apply(lambda dictionary : dictionary[coln] if not pd.isna(dictionary) and coln in dictionary.keys() else np.NaN)
    return outputDF
```

```python
hmdb_xrefs = atomiclize_col(data['xrefs'])[['xrefs_inchikey', 'xrefs_chemspider', 'xrefs_kegg_compound', 'xrefs_drugbank', 'xrefs_foodb', 'xrefs_pdb', 'xrefs_chebi', 'xrefs_pubchem_cid', 'xrefs_smiles']]
hmdb_xrefs
```

```python
hmdb_inchikeyTo_microbes = pd.DataFrame(data['associated_microbes'].explode()).merge(hmdb_xrefs[['xrefs_inchikey']], left_index = True, right_index=True)


# expand every column that has dictionary as its values
for col in hmdb_inchikeyTo_microbes.columns:
    if col != 'xrefs_inchikey':
        uncleaned_col = hmdb_inchikeyTo_microbes[col]
        hmdb_inchikeyTo_microbes.drop(columns = [col], inplace = True)
        hmdb_inchikeyTo_microbes = pd.concat([hmdb_inchikeyTo_microbes, atomiclize_col(uncleaned_col)], axis=1)

hmdb_inchikeyTo_microbes
```

```python
hmdb_inchikeyTo_pathways = pd.DataFrame(data['associated_pathways'].explode()).merge(hmdb_xrefs[['xrefs_inchikey']], left_index = True, right_index=True)

for col in hmdb_inchikeyTo_pathways.columns:
    if col != 'xrefs_inchikey':
        uncleaned_col = hmdb_inchikeyTo_pathways[col]
        hmdb_inchikeyTo_pathways.drop(columns = [col], inplace = True)
        hmdb_inchikeyTo_pathways = pd.concat([hmdb_inchikeyTo_pathways, atomiclize_col(uncleaned_col)], axis=1)
        
hmdb_inchikeyTo_pathways
```

```python
hmdb_inchikeyTo_proteins = pd.DataFrame(data['associated_proteins'].explode()).merge(hmdb_xrefs[['xrefs_inchikey']], left_index = True, right_index=True)

for col in hmdb_inchikeyTo_proteins.columns:
    if col != 'xrefs_inchikey':
        uncleaned_col = hmdb_inchikeyTo_proteins[col]
        hmdb_inchikeyTo_proteins.drop(columns = [col], inplace = True)
        hmdb_inchikeyTo_proteins = pd.concat([hmdb_inchikeyTo_proteins, atomiclize_col(uncleaned_col)], axis=1)
        
hmdb_inchikeyTo_proteins
```
