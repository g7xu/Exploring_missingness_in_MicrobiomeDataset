```python
import pandas as pd
import numpy as np
from pathlib import Path  
```

```python
data = pd.read_json('data/disbiome_data.json')
data
```

```python
def atomiclize_col(col):
    '''
    expand any columns with complex dictionaries
    ---
    col: takes in a series where each entry is a dictionary
    ---
    return a DataFrame with expanded 
    '''
    key_collection = set()
    for i in col.values:
        if not pd.isna(i):
            key_collection = key_collection.union(set(i.keys()))
    outputDF = pd.DataFrame()
    for coln in key_collection:
        outputDF[col.name+ '_' + coln] = col.apply(lambda dictionary : dictionary[coln] if not pd.isna(dictionary) and coln in dictionary.keys() else np.NaN)
    return outputDF
```

```python
for col in data.columns:
    if col != '_id':
        uncleaned_col = data[col]
        data.drop(columns = [col], inplace = True)
        data = pd.concat([data, atomiclize_col(uncleaned_col)], axis=1)
data
```
