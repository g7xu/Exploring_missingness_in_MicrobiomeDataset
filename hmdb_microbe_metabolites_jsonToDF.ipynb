```python
hmdb = data[['_id', 'name', 'chemical_formula', 'description', 'status', 'average_mw', 'monoisotopic_mw', 'state', 'sources']]
hmdb = hmdb.assign(inchikey = data['xrefs'].apply(lambda x : x['inchikey']))
hmdb
```

```python
def atomiclize_col(col):
    '''
    expand any columns with complex dictionaries
    ---
    col: takes in a series where each entry is a dictionary
    ---
    return a DataFrame with expanded 
    '''
    key_collection = set()
    for i in col.index:
        key_collection = key_collection.union(set(col.loc[i].keys()))
    outputDF = pd.DataFrame()
    for coln in key_collection:
        outputDF[col.name+ '_' + coln] = col.apply(lambda dictionary : dictionary[coln] if coln in dictionary.keys() else np.NaN)
    return outputDF
```

```python
hmdb_xrefs = atomiclize_col(data['xrefs'])[['xrefs_inchikey', 'xrefs_chemspider', 'xrefs_kegg_compound', 'xrefs_drugbank', 'xrefs_foodb', 'xrefs_pdb', 'xrefs_chebi', 'xrefs_pubchem_cid', 'xrefs_smiles']]
hmdb_xrefs
```
